# Checking variance and removing columns with 0 variance

```python
threshold = 0.0

traindf=traindf.drop(traindf.var()[traindf.var() <= threshold].index.values, axis=1)
```

```python
for c in traindf.columns:
    if traindf[c].dtype == 'object':
        lbl = LabelEncoder()
        lbl.fit(list(traindf[c].values) + list(testdf[c].values))
        traindf[c] = lbl.transform(list(traindf[c].values))
        testdf[c] = lbl.transform(list(testdf[c].values))
```

# Splitting the data

```python
traindf=traindf.drop('ID',axis=1)
testdf=testdf.drop('ID',axis=1)
```

```python
X_train=traindf.drop('y',axis=1)
y_train=traindf['y']
X_test=testdf
```

# PCA for dimensionality reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
X_trainpca=pca.fit_transform(X_train)
X_testpca=pca.fit_transform(X_test)
print(pca.explained_variance_ratio_.sum())
print("--------------------------------------------------------------")
print(pca.components_)
print("--------------------------------------------------------------")
print(pca.components_.shape)
print("--------------------------------------------------------------")
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
pd.DataFrame(X_testpca).head()
```

# Regression using XGBOOST

```python
X_train=pd.DataFrame(X_trainpca)
y_train=pd.DataFrame(traindf['y'])
X_test=pd.DataFrame(X_testpca)
dtrain = xgb.DMatrix(X_train, label=y_train)
xgb_model = xgb.XGBRegressor(base_score=0.05, colsample_bylevel=1,
                   colsample_bynode=1, colsample_bytree=0.5, 
                   learning_rate=0.2, gamma=0.0001, importance_type='gain',
                   max_depth=4, n_estimators=200,reg_alpha=0., reg_lambda=1, subsample=0.5, 
                   objective='reg:squarederror', random_state=28)
#(objective="reg:squarederror", random_state=42,silent=True,base_score=0.1,                             eval_metric='rmse',eta=0.01)#booster='gblinear'

xgb_model.fit(X_train, y_train)
train_pred = xgb_model.predict(X_train)

mse=mean_squared_error(y_train, train_pred)


# Accuray Score on train dataset
#accuracy_train = score.xgb.model(y_train,train_pred)
print('R2 : ', r2_score(y_train,train_pred))

#RMSE of the train dataset
print('RMSE :',np.sqrt(mse))

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
kf = KFold(shuffle=True, n_splits=15)

print('Accuracy of Model with Cross Validation is:',cross_val_score(xgb_model, X_train, y_train, cv=kf,scoring='r2').mean()*100)
```

# From the value of the R2(0.48130219343452896) ,we can say that the model performance is very low.As we know that a good R2 should be near to 1.
